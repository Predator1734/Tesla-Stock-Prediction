```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

from matplotlib import style
style.use("fivethirtyeight")
```

```python
# Load the datset
df = pd.read_csv(r"Tesla.csv")
```

```python
# Displaying first 5 rows of the dataset
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Creating two columns for prediction
df['HL_Perc'] = (df['High']-df['Low']) / df['Low'] * 100
df['CO_Perc'] = (df['Close']-df['Open']) / df['Open'] * 100
```

```python
dates = np.array(df['Date']) #converts the dates column into a numpy array
dates_check = dates[-30:]  # creates new column with last 30 dates
dates = dates[:-30] # updates the dates column excluding the last 30 dates
```

```python
df = df[['HL_Perc','CO_Perc','Adj Close','Volume']]
```

```python
# define the label column
df['PriceNextMonth'] = df['Adj Close'].shift(-30)
```

```python
df.tail()
```

```python
# Make feature and label arrays
X = np.array(df.drop(['PriceNextMonth'],axis=1))
X = preprocessing.scale(X)
X_Check = X[-30:]
X = X[:-30]
df.dropna(inplace=True)
y = np.array(df['PriceNextMonth'])
```

```python
# Splitting the data into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# introducing Random Forest Algorithm
model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
# Claculating the confidence value by aplying the model to testing data
conf = model.score(X_test,y_test)
print('Confidence Value :',conf)
```

```python
# fit the model using the whole dataset
model.fit(X,y)
```

```python
predictions = model.predict(X_Check)
```

```python
# make the final dataframe containing Dates, ClosePrices and Forcast Values
actual = pd.DataFrame(dates, columns = ['Date'])
actual['ClosePrice']=df['Adj Close']
actual['Forecast'] = np.nan
actual.set_index('Date',inplace=True)
forecast = pd.DataFrame(dates_check, columns = ['Date'])
forecast['Forecast'] = predictions
forecast['ClosePrice'] = np.nan
forecast.set_index('Date', inplace=True)
var = [actual, forecast]
result = pd.concat(var) #final dataframe
```

```python
#plottng the final result
result.plot(figsize=(20,10), linewidth = 1.5)
plt.legend(loc = 2, prop={'size':20})
plt.xlabel("Date")
plt.ylabel("Price")
```
